```python
chrome_service = Service(executable_path='chromedriver-mac-arm64/chromedriver')
driver = webdriver.Chrome(service=chrome_service)
```

```python
FullName = []
Location = []
Sector = []
Position = []
Started = []
HBYear = []
firstEdu = []
latestEdu = []
#thirdEdu = []




for i in searchList:
    print(i)
    #open page
    try:
        driver.get(i)
        time.sleep(5)

        page_source = driver.page_source

        soup = BeautifulSoup(page_source, 'html.parser')
        name = soup.find('h1', {'class': 'top-card-layout__title font-sans text-lg papabear:text-xl font-bold leading-open text-color-text mb-0'}).text.strip()
        Industry = soup.find('h2', 
                             {'class': 'top-card-layout__headline break-words font-sans text-md leading-open text-color-text'}
                            ).text.strip().split(', ')[-1]

        countryLocation = soup.find('div', 
                             {'class': 'not-first-middot'}
                            ).find('span').text.strip()

        fullExperience = soup.find_all('li', 
                                  {'class': 'profile-section-card relative flex w-full list-none py-1.5 pr-2 pl-1 experience-item'}
                                  )

        
        Current_Position = fullExperience[1].find('div').find('h3').text.strip()
        Current_Insitution = fullExperience[1].find('div').find('h4').text.strip()
        yearStarted = fullExperience[1].find('div').find('div').find('p').find('span').find('time').text.strip()
        
        #education
        allEducation = soup.find('ul', {'class': 'education__list'}).find_all('li')
        educationOne = allEducation[-1].find('div').find('h3').text.strip()
        educationTwo = allEducation[1].find('div').find('h3').text.strip()
        #educationThree = allEducation[2].find('div').find('h3').text.strip()

        

        FullName.append(name)
        Location.append(countryLocation)
        Position.append(Current_Position)
        Sector.append(Industry)
        Started.append(yearStarted)
        
        firstEdu.append(educationOne)
        latestEdu.append(educationTwo)
        #thirdEdu.append(educationThree)
        
        
        for exp in fullExperience:
            pattern = re.compile(r'HackBio')
            match = re.search(pattern, str(exp))
            
            if match:
                yearHB = exp.find('div').find('div').find('p').find('span').find('time').text.strip()
                
                HBYear.append(yearHB)
                #fullExperience.next()
            else:
                HBYear.append(None)
            
        
    except:
        
        FullName.append(None)
        Location.append(None)
        Position.append(None)
        Started.append(None)
        Sector.append(None)
        HBYear.append(None)
        firstEdu.append(None)
        latestEdu.append(None)
        #thirdEdu.append(None)


```

```python
outTab = pd.DataFrame({'FullName': FullName,
                      'Current_Location': Location,
                      'Current_Position': Position,
                      'Started': Started,
                      'HBYear': HBYear,
                      'Sector': Sector})

outTab.head()
#FullName
#Location
#Position
#Started
```
